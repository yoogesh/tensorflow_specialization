```python
import numpy as np
import tensorflow as tf 
import keras as k
from keras import Sequential
```

```python
#creating the data 
x = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype = float)
t = np.array([-3.0, -1.0, 1.0, 3.0, 5., 7.0], dtype = float)
```

```python
# creating an tensorflow model 
model_1 = k.Sequential([k.layers.Dense(units = 1, input_shape = [1])])
```

```python
model_1.compile(optimizer = 'sgd', loss = 'mean_squared_error')
```

```python
model_1.fit(x, t, epochs = 500)
```

```python
model_1.predict([20])
```
